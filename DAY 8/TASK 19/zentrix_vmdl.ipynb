```python
model = YOLO("yolov8n.pt")
```

```python
results = model.train(
    data="/content/ppe_dataset/data.yaml",
    epochs=20,
    imgsz=640,
    batch=8,
    patience=5,
    augment=True,
    workers=2,
    cache=True,
    save=True,
    project="/content/gdrive/MyDrive/Zentrix_AI",
    name="ppe_detector"
)
```

```python
metrics = model.val()
print("mAP@0.5 :", round(metrics.box.map50, 3))
print("Precision :", round(metrics.box.mp, 3))
print("Recall :", round(metrics.box.mr, 3))
```

```python
results = model.predict(
    source="/content/0QHQ2nmdE78_clip0_trim_7.mp4",
    conf=0.4
)
violations = []
for r in results:
    for box in r.boxes:
        cls = int(box.cls[0])
        label = model.names[cls]
        if "no_" in label.lower():
            violations.append(label)
print("Violations Detected:")
print(set(violations))
```

```python
if len(violations) >= 3:
    risk = "HIGH"
elif len(violations) >= 1:
    risk = "MEDIUM"
else:
    risk = "LOW"
print("Predicted Risk Level :", risk)
```
